```python
%matplotlib inline

from Data_Preprocessing import Data_Preprocessing
from MLP_Model import MLP_Model
from collections import OrderedDict
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import datasets
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib.gridspec as gridspec
import itertools

TRAIN_FILE_PATH = 'train_file_path.txt'

TEST_FILE_PATH = 'test_file_path.txt'

LABEL_NAMES = ['Carambula', 'Lychee', 'Pear']

EPOCHS = 4000

FULL_BATCH = 1470

BATCH_SIZE = 128

# N_HIDDENS = [512]
N_HIDDENS = [32, 32]

LEARNING_RATE = 0.01

# ACTIVATION_FUNC = 'RELU'

ACTIVATION_FUNC = 'SIGMOID'

KEEP_RATE = 0.5

def get_random_color_dict(n_labels):
    color_dict = {}
    RGB = 3
    for i in range(n_labels):
        while(True):
            color = [random.random() for _ in range(RGB)]
            if color not in color_dict.values():
                color_dict[i] = color
                break
    return color_dict
        
def visualize_data_in_2D(data, label, color_dict, title_name):    
    plt.title(title_name)
    for i in range(data.shape[0]):        
        color = color_dict[int(label[i])]
        plt.scatter(data[i, 0], data[i, 1], color=color, label=LABEL_NAMES[label[i]])
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), loc='upper right')
    plt.show()

def plot_labels_with_colors(data, one_hot_label, one_hot_label_hat):
    pca_data = Data_Preprocessing.get_PCA_data(data=data, n_components=2)
    color_dict = get_random_color_dict(n_labels=one_hot_label.shape[1])
    visualize_data_in_2D(data=pca_data, label=np.argmax(one_hot_label, axis=1), color_dict=color_dict, title_name='true label')
    visualize_data_in_2D(data=pca_data, label=np.argmax(one_hot_label_hat, axis=1), color_dict=color_dict, title_name='predict_label')

def _plot_decision_region(data, label, title_name):
    clf1 = LogisticRegression(random_state=1, solver='newton-cg', multi_class='multinomial')                              
    clf2 = RandomForestClassifier(random_state=1, n_estimators=100)
    clf3 = GaussianNB()
    clf4 = SVC(gamma='auto')
    gs = gridspec.GridSpec(2, 2)    
    fig = plt.figure(figsize=(10,8))
    labels = ['Logistic Regression', 'Random Forest', 'Naive Bayes', 'SVM']
    for clf, lab, grd in zip([clf1, clf2, clf3, clf4], labels, itertools.product([0, 1], repeat=2)):                                
        clf.fit(data, label)
        ax = plt.subplot(gs[grd[0], grd[1]])
        fig = plot_decision_regions(X=data, y=label, clf=clf, legend=2)
        plt.title(lab+' ('+ title_name + ')')   
        handles, labels = ax.get_legend_handles_labels()
        ax.legend(handles, LABEL_NAMES, framealpha=0.3, scatterpoints=1)
    plt.show()
    
def plot_decision_region(data, one_hot_label, one_hot_label_hat):
    pca_data = Data_Preprocessing.get_PCA_data(data=data, n_components=2)
    _plot_decision_region(data=pca_data, label=np.argmax(one_hot_label, axis=1), title_name='true label')
    _plot_decision_region(data=pca_data, label=np.argmax(one_hot_label_hat, axis=1), title_name='predict label')


if __name__ == '__main__':

    train_dp = Data_Preprocessing(file_path=TRAIN_FILE_PATH,
                                  label_names=LABEL_NAMES)

    test_dp = Data_Preprocessing(file_path=TEST_FILE_PATH,
                                 label_names=LABEL_NAMES)

    train_data, train_label = train_dp.get_data_and_label(data_process=False,
                                                          one_hot_label=True)

    test_data, test_label = test_dp.get_data_and_label(data_process=False,
                                                       one_hot_label=True)

    model = MLP_Model(n_features=train_data.shape[1],
                      n_hiddens=N_HIDDENS,
                      n_labels=len(LABEL_NAMES),
                      activation_func=ACTIVATION_FUNC,
                      learning_rate=LEARNING_RATE)

    model.fit(train_data=train_data,
              train_label=train_label,
              test_data=test_data,
              test_label=test_label,
              epochs=EPOCHS,
              batch_size=BATCH_SIZE,
             keep_rate=KEEP_RATE)

    test_label_hat, _ = model.predict(input_data=test_data,
                                                  one_hot_label=test_label)
    

    plot_labels_with_colors(data=test_data,
                                     one_hot_label=test_label,
                                     one_hot_label_hat=test_label_hat)

    plot_decision_region(data=test_data,
                                  one_hot_label=test_label,
                                  one_hot_label_hat=test_label_hat)


```
